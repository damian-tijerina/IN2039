```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import statsmodels.api as sm

LG=pd.read_excel('Leche CAETEC (1).xlsx',
                sheet_name='Periodo 2021',
                 fill_value=0,
                engine='openpyxl')
LG.head()
```

# Regresión (ST, P,G Y L)

```python
X = LG['Grasa'].values
Y = LG['Sólidos Totales'].values
print(X)
print(Y)
```

```python
np.corrcoef(X, Y)
```

```python
model = sm.OLS.from_formula('Y~X', # y= b0 + b1x   modelo de regresión simple osea una sola variable
                           data=LG)

result = model.fit()
result.summary()
```

# Colonias

```python
LG.shape
```

```python
data= LG.sample(n=351)
```

```python
line_datag = data.groupby('Fecha')['Colonias'].mean().reset_index()
line_datag
```

```python
fig = go.Figure(data=go.Scatter(x=line_datag['Fecha'],
                               y=line_datag['Colonias'],
                               mode='lines',
                               marker=dict(color='green')
                               )
               )
fig.show()
```
