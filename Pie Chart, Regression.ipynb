```python
import numpy as no
import pandas as pd
df=pd.read_excel('datasets/Canada.xlsx',
               sheet_name='Canada by Citizenship',
               skiprows=20,
               skipfooter=2,
               engine='openpyxl')
df
```

```python
df.columns
```

```python
df.drop(['Type',    'Coverage',  'AREA', 'REG', 'DEV'],
       axis=1,
       inplace=True)
df.rename(columns={'OdName':'Country',
                  'AreaName':'Continent',
                  'RegName':'Region'},
         inplace=True)
```

```python
df.head()
```

```python
df.columns=list(map(str,df.columns))
df.set_index('Continent', inplace=True)
years=list(map(str,range(1980,2014)))

new_df= df[years]
new_df
```

```python
new_df = new_df.dropna(how='all')
```

```python
new_df
```

```python
new_df=new_df[years]
new_df['Total']=new_df.iloc[:,:].sum(axis=1)
new_df
```

```python
new_df= new_df.groupby('Continent',
                      axis=0).sum()
new_df
```

```python
new_df.drop(['World'],
           axis=0,
           inplace=True)
new_df
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
color_list=['gold','yellowgreen','lightcoral','lightskyblue','lightgreen','pink']
exp_list=[0.1, 0.1, 0, 0, 0.2, 0.3] #Es para remarcar cuales son las que tienen números más bajos


new_df['Total'].plot(kind='pie',
                    figsize=(15,6),
                    autopct= '%10.1f%%',
                    startangle = 90,
                    shadow = True,
                    labels=None,
                    colors= color_list,
                    explode = exp_list)
plt.title('Inmigration to Canada by Continent 1980-2013')
plt.axis('equal')
plt.legend(labels=new_df.index, loc='upper left') #Cuadro con países
plt.show()
```

```python
#Suma de las personas por años
df_t = pd.DataFrame(new_df[years].sum(axis =0))
df_t.head()
```

```python
df_t.index= map(float, df_t.index)
df_t.reset_index(inplace=True)
df_t.head(2)
```

```python
df_t.columns=['years','total']
df_t.head(2)
```

```python
import seaborn as sns
mpl.style.use('ggplot')
plt.figure(figsize = (15,10))
sns.set(font_scale =1.5)

reg_plot = sns.regplot (x='years',
                       y='total',
                       data = df_t,
                       color = 'm',
                       marker='+',
                       scatter_kws = {'s':200})

reg_plot.set(xlabel = 'Year',
      ylabel = 'Total immigration')
reg_plot.set_title('Total immigration to Canada 1980 - 2013')
plt.show()
```
