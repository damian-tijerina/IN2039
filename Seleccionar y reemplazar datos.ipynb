```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 100)
df = pd.read_csv('datasets/nhanes_2015_2016.csv')
df.head()
```

# Take dataframe from only certain columns

```python
col_names= df.columns
col_names
```

```python
[column for column in col_names if 'BMX' in column]
#One way to get only column names that include 'BMX' is with list comprehension
# [keep x for x in list if condition met]
```

```python
keep = [column for column in col_names if 'BMX' in column]
keep
```

```python
df[keep]
```

```python
df_BMX=df[keep]
df_BMX.head()
```

# Seleccionar datos con .loc o .iloc

```python
df.loc[:,keep].head()
```

```python
#Identifies which columns have the keep condition by putting True or False
index_bool = np.isin(df.columns,keep)
index_bool
```

```python
#If iloc will be used then it´s necessary transform columns into index
df.iloc[:,index_bool].head()
```

```python
df.loc[:,index_bool].head()
```

# Selección de datos con condiciones

```python
df_BMX.dtypes
```

```python
waist_median = pd.Series.median(df_BMX['BMXWAIST'])
waist_median
```

```python
waist_mean = pd.Series.mean(df_BMX['BMXWAIST'])
waist_mean
```

```python
df_BMX[df_BMX['BMXWAIST']> waist_median].head()
```

```python
#Add conditions
condition1 = df_BMX['BMXWAIST'] > waist_median
condition2 = df_BMX['BMXLEG'] < 32

df_BMX[condition1 & condition2].head()
```

```python
df_BMX.loc[condition1 & condition2, :].head() 
#The .loc method is describing the rows to keep
```

```python
tmp = df_BMX.loc[condition1 & condition2, :].head()
tmp.index = ['Primera','Segunda','Tercera','Cuarta','Quinta']
tmp
```

```python
tmp.loc[['Primera','Segunda'], 'BMXLEG']
```

```python
tmp.iloc[[0,1],3]
```
