```python
import pandas as pd
import numpy as np
df=pd.read_csv('datasets/headbrain.csv')
df.head(2)
```

```python
X = df['Head Size(cm^3)'].values
Y = df['Brain Weight(grams)'].values
print(X)
print(Y)
```

```python
#Ver si hay una correlación entre las dos variables de arriba
np.corrcoef(X, Y)
#Matriz de correlación
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,Y,
           c='y',
           label='Data points')
plt.xlabel('Head Size(cm^3)')
plt.ylabel('Brain Weight(grams)')

plt.legend()
plt.show()
```

```python
import statsmodels.api as sm

model = sm.OLS.from_formula('Y~X', # y= b0 + b1x   modelo de regresión simple osea una sola variable
                           data=df)

result = model.fit()
result.summary()

#Arroja la estadística que un software como minitab arroja para regresión lineal
```

```python
from sklearn import linear_model
lm= linear_model.LinearRegression()
X= pd.DataFrame(df['Head Size (cm^3)'])
Y= pd.DataFrame(df['Brain Weight(grams)'])
model_lm= lm.fit(X,Y)
model_lm
```
